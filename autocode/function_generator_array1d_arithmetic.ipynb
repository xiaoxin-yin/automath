```python
import numpy as np
import random

def generate_random_array():
    length = random.randint(2, 8)  # Randomly choose the length of the array
    # Generate the array with absolute values following an exponential distribution
    array = [(-1)**random.randint(0, 1) * round(np.random.exponential(scale=8.0)) for _ in range(length)]
    return array

random_array = generate_random_array()
print("Random Array:", random_array)
```

```python
import random

def square(x):
    return x*x

def generate_random_operator():
    operators = ['+', '-', '*', '//', '%', 'max', 'min']
    return random.choice(operators)

def generate_random_comparison():
    comparisons = ['<', '>', '<=', '>=', '==', '!=']
    return random.choice(comparisons)

def generate_random_unary_function():
    unary_functions = ['abs', 'square']
    return random.choice(unary_functions)

def generate_random_constant(arr_length=False, positive_only=False):
    if arr_length and random.choice([True, False]):
        unary_functions = ['']  # This is useless. Should put at base_case_constant
        unary_function = random.choice(unary_functions)
        if unary_function:
            return random.choice([f'{unary_function}(foo(arr[1:]))'])  # f'{unary_function}(len(arr))', 
        else:
            return random.choice(['len(arr)', 'foo(arr[1:])'])
    else:
        return str(random.randint(0 if positive_only else -5, 10))

def generate_random_code():
    base_case_constant = generate_random_constant()
    positive_constant = generate_random_constant(positive_only=True)
    recursive_case_constant = generate_random_constant(arr_length=True)
    base_case_operator = generate_random_operator()
    recursive_operator = generate_random_operator()
    comparison_operator = generate_random_comparison()
    unary_function = generate_random_unary_function()

    base_case_code = "arr[0]"
    rand_num = random.random()
    if rand_num < 0.33:
        if base_case_operator in ['max', 'min']:
            base_case_code = f"{base_case_operator}([arr[0], {base_case_constant}])"
        else:
            base_case_code = f"(arr[0] {base_case_operator} {base_case_constant})"
    elif rand_num < 0.66:
        base_case_code = f"{unary_function}(arr[0])"

    rand_num = random.random()
    recursive_part = f"foo(arr[1:])"
    if rand_num < 0.167:
        recursive_part = f"foo(arr[1:])"
    if rand_num < 0.333:
        recursive_part = f"arr[0]"
    elif rand_num < 0.5:
        recursive_part = f"{base_case_code}"
    else:
        if recursive_operator in ['max', 'min']:
            recursive_part = f"{recursive_operator}([{base_case_code}, {recursive_part}])"
        else:
            recursive_part = f"{base_case_code} {recursive_operator} {recursive_part}"

    if random.random()>0.7:
        code = f"""def foo(arr):
    if len(arr) == 1:
        return {base_case_code}
    else:
        if len(arr) {comparison_operator} {positive_constant}:
            return {recursive_part}
        else:
            return foo(arr[1:])
    """  
    else:
        code = f"""def foo(arr):
    if len(arr) == 1:
        return {base_case_code}
    else:
        return {recursive_part}
    """

    return code

# Generate random code
random_code = generate_random_code()
print(random_code)

# Example usage:
exec(random_code)
random_array = generate_random_array()
print("Random Array:", random_array)
print("Result:", foo(random_array))
```
