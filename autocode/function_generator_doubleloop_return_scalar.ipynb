```python
import numpy as np
import random

def generate_random_array():
    length = random.randint(2, 8)  # Randomly choose the length of the array
    # Generate the array with absolute values following an exponential distribution
    array = [(-1)**random.randint(0, 1) * round(np.random.exponential(scale=8.0)) for _ in range(length)]
    return array

random_array = generate_random_array()
print("Random Array:", random_array)
```

```python
import random

def generate_random_double_loop_function():
    operators = ['+', '-', '*', '//', 'append', 'add']
    functions = ['max', 'min', 'sum', 'len']
    comparisons = ['>', '<', '>=', '<=', '==', '!=']
    state_types = ['int', 'list', 'set']
    int_initial_values = {
        '+': '0',
        '-': '0',
        '*': '1',
        '//': '1',
        'append': 'float(\'-inf\')',
        'add': 'float(\'-inf\')'
    }

    template = """
def random_function(arr):
    best_state = {init_best_state}
    for i in range(len(arr)):
        current_state = {init_current_state}
        for j in range(i, len(arr)):
            if '{op}' in ['append', 'add']:
                current_state.{op}(arr[j])
            else:
                current_state = current_state {op} arr[j]
            if (isinstance(current_state, (list, set)) and len(current_state) {comp} len(best_state)) or (isinstance(current_state, int) and current_state {comp} best_state):
                best_state = {func}(best_state, current_state) if isinstance(best_state, int) and isinstance(current_state, int) else current_state
    return best_state
"""

    # Randomly select an operator, function, comparison, and state type
    selected_op = random.choice(operators)
    selected_func = random.choice(functions)
    selected_comp = random.choice(comparisons)
    selected_state_type = random.choice(state_types)

    # Determine initial values based on the selected state type, function, and operator
    if selected_state_type == 'int':
        init_best_state = 'float(\'inf\')' if selected_func == 'min' else 'float(\'-inf\')'
        init_current_state = int_initial_values[selected_op]
    else:
        init_best_state = '[]' if selected_state_type == 'list' else 'set()'
        init_current_state = '[]' if selected_state_type == 'list' else 'set()'

    # Fill in the placeholders in the template
    function_code = template.format(
        op=selected_op,
        func=selected_func,
        comp=selected_comp,
        init_best_state=init_best_state,
        init_current_state=init_current_state
    )

    return function_code

# Generate a random double-loop function
random_function_code = generate_random_double_loop_function()
print(random_function_code)

# You can then execute this code using exec() or save it to a file for later use.
```
