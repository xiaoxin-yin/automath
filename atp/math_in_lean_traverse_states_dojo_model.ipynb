```python
import re

def parse_lean_definition(lean_code):
    # Regular expression to capture the name, signature, and content
    pattern = r"\bdef\s+(\w+)\s*(.*?)\s*:=\s*(.*)"
    match = re.search(pattern, lean_code, re.DOTALL)
    
    if match:
        name = match.group(1).strip()
        signature = match.group(2).strip()
        content = match.group(3).strip()
        if signature.endswith('Prop'):
            signature = signature[:-4].strip()
        if signature.endswith(':'):
            signature = signature[:-1].strip()
        return name, signature, content
    else:
        # Regular expression to capture the name, signature, and content
        pattern = r"\bdef\s+(\w+)\s*(.*)"
        match = re.search(pattern, lean_code, re.DOTALL)
        if match:
            name = match.group(1).strip()
            return name, '', ''
        return None, None, None

# # Example usage
# lean_code = "def FnLb (f : ℝ → ℝ) (a : ℝ) : Prop :=\n  ∀ x, a ≤ f x"
# result = parse_lean_definition(lean_code)
# if result:
#     name, signature, content = result
#     print("Name:", name)
#     print("Signature:", signature)
#     print("Content:", content)
# else:
#     print("No match found.")


def get_custom_definitions(traced_file):
    premises = traced_file.get_premise_definitions()
    definitions = {}
    for premise in premises:
        if premise['code'].startswith("def "):
            definition = parse_lean_definition(premise['code'])
            if definition != None:
                definitions[definition[0]] = definition[1]
    return definitions

get_custom_definitions(traced_file)
```

```python
import lean_dojo
from random import randint
import time


from utils.lean_math_utils import *

MAX_STEPS = 100000

theorem = Theorem(repo, "MIL/C02_Basics/solutions/Solutions_S05_Proving_Facts_about_Algebraic_Structures.lean",
                  "Solutions_S05_Proving_Facts_about_Algebraic_Structures_ex1")

def explore_state_complexity(state):
    complexity = 0
    lines = state.pp.split('\n')
    targets = []
    for line in lines:
        if line.startswith("⊢"):
            targets.append(line[1:].strip())
    lengths = [len(x) for x in targets]
    complexity = max(lengths)/2 + sum(lengths)/2
    return complexity

# Print the trace from state_0 to state
def get_tactic_trace(curr_state, state_dict):
    _, parent_states, tactics = state_dict[curr_state]
    if len(parent_states) == 0:
        return [(tactic, curr_state)]
    else:
        return get_tactic_trace(parent_states[0], state_dict) + [(tactic, curr_state)]

def explore_states(theorem, tac_templates, theorem_code=None, max_steps=MAX_STEPS, max_time=1200, verbose=False):
    start_time = time.time()
    # For some theorems, it might take a few minutes.
    dojo, state_0 = Dojo(theorem).__enter__()

    #if verbose: print(state_0.pp)
    unused_vars = []
    if theorem_code is not None:
        tokens = tokenize_lean_tactic(theorem_code)
        tokens = [x for x in tokens if x.strip() != '']
        type_of_item, def_of_item = classify_lean_elements(state_0.pp)
        unused_vars = [x for x in type_of_item.keys() if x not in tokens]
        #print(unused_vars)

    state_queue = PriorityQueue() # PQ of states, priority being complexity of the state
    state_dict = {} # Keyed by state.pp. Value is (state, its parent states, and tactics from parent state to this state)

    curr_state = state_0
    state_queue.push(curr_state, explore_state_complexity(curr_state) + randint(0, 4))
    state_dict[curr_state] = (curr_state, list(), list())

    n_steps = 0
    theorem_proven = False
    while state_queue.size() > 0:
        curr_state = state_queue.pop()
        if not hasattr(curr_state, 'pp'):
            continue
        #print("POP STATE complexity", explore_state_complexity(curr_state), curr_state.pp, "\nEND STATE")
        type_of_item, def_of_item = classify_lean_elements(curr_state.pp)
        for var in unused_vars:
            del type_of_item[var]
            del def_of_item[var]
        if verbose:
            if n_steps == 0:
                print("INIT_STATE:", curr_state.pp)
                print(type_of_item)
                #print(def_of_item)
        #
        #TODO: Add {'nvar0': 'nvar0', 'nvar1': 'nvar1', ...} to type_of_item
        #
        tactic_set = set()
        for tac_template in tac_templates:
            tactics = generate_tactics_from_template(tac_template, type_of_item)
            for tactic in tactics:
                tactic_set.add(tactic)
        #
        proof_finished = False
        tactic_trace = None
        #print(len(tactic_set), "tactics")
        for tactic in tactic_set:
            test_state = dojo.run_tac(curr_state, tactic)
            n_steps += 1
            if n_steps % 10000 == 0:
                print(n_steps, "steps executed")
            if n_steps > max_steps:
                break
            if type(test_state) == lean_dojo.interaction.dojo.LeanError:
                continue
            elif type(test_state) == lean_dojo.interaction.dojo.ProofFinished:
                proof_finished = True
                if verbose:
                    print("ProofFinished")
#                     tactic_trace = get_tactic_trace(curr_state, state_dict) + [(tactic, test_state)]
#                     for i in range(len(tactic_trace)):
#                         tactic, state = tactic_trace[i]
#                         print("STEP", i)
#                         if tactic != None:
#                             print("Tactic:", tactic)
#                         if hasattr(state, 'pp'):
#                             print(state.pp)
                if test_state in state_dict:
                    _, parent_states, tactics = state_dict[test_state]
                    state_dict[test_state] = (test_state, parent_states + [curr_state], tactics + [tactic])
                else:
                    complexity = 100000000
                    state_queue.push(test_state, complexity)
                    state_dict[test_state] = (test_state, [curr_state], [tactic])
            else:
                #print("TEST STATE:", test_state.pp)
                if test_state in state_dict:
                    _, parent_states, tactics = state_dict[test_state]
                    state_dict[test_state] = (test_state, parent_states + [curr_state], tactics + [tactic])
                else:
                    complexity = explore_state_complexity(test_state)
                    state_queue.push(test_state, complexity  + randint(0, 4))
                    state_dict[test_state] = (test_state, [curr_state], [tactic])
        #
        if proof_finished:
            theorem_proven = True
            #break
        if n_steps > max_steps:
            break     
        cur_time = time.time()
        if (cur_time - start_time) > max_time:
            print("Max run-time exceeded")
            break
    #
    return state_dict, theorem_proven, tactic_trace

state_dict, theorem_proven, tactic_trace = \
    explore_states(theorem,
                   tac_templates,
                   theorem_code='Solutions_S05_Proving_Facts_about_Algebraic_Structures_ex1 : x ⊓ y = y ⊓ x', 
                   max_steps = 100000,
                   verbose=True)

print(theorem_proven)
print(tactic_trace)
```

```python
# Generate state pairs and their distances. 
# Probably useful if we hope to generate intermediate states (e.g., "have").

def yield_state_pairs(curr_state, leaf_state, state_dict, seen_states_pp, max_distance=8, max_parents=10):
    #     print("curr_state:", curr_state.pp)
    #     print("leaf_state:", leaf_state.pp)
    _, parent_states, tactics = state_dict[curr_state]
    if len(seen_states_pp) > 1:
        yield (curr_state, leaf_state, len(seen_states_pp) - 1)
    if len(seen_states_pp) <= max_distance:
        num_parents_checked = 0
        for parent_state in parent_states:
            if parent_state.pp not in seen_states_pp:
                yield from yield_state_pairs(parent_state, leaf_state, state_dict, seen_states_pp + [parent_state.pp])
                num_parents_checked += 1
                if num_parents_checked >= max_parents:
                    break
    return


def get_state_provability_data(theorem, tac_templates, theorem_code=None, max_steps=100000, negative_ratio=1.0, verbose=False):
    state_dict, theorem_proven, tactic_trace = \
    explore_states(theorem,
                   tac_templates,
                   theorem_code=theorem_code, 
                   max_steps=max_steps,
                   verbose=verbose)
    #
    all_states = list(state_dict.values())
    print(len(all_states), "states in total")
    proven_states = [x[0] for x in all_states if type(x[0]) == lean_dojo.interaction.dojo.ProofFinished]
    print(len(proven_states), "proven states")
    state_pairs = []
    seen_state_pps = set()
    #
    for proven_state in proven_states:
        print(proven_state)
        pairs = list(yield_state_pairs(proven_state, proven_state, state_dict, [str(proven_state)]))
        ancestor_state_dict = {}
        for ancestor, leaf, distance in pairs:
            if ancestor.pp not in ancestor_state_dict:
                ancestor_state_dict[ancestor] = distance
            else:
                ancestor_state_dict[ancestor] = min(distance, ancestor_state_dist[ancestor])
        #print(ancestor_state_dist)
        print(len(ancestor_state_dict), "ancestors found")
        for ancestor, distance in ancestor_state_dict.items():
            state_pairs.append((ancestor, proven_state, distance))
            seen_state_pps.add(ancestor.pp)
    #
    num_positive = len(state_pairs)
    num_negative = 0
    for state in all_states:
         if hasattr(state, 'pp') and state.pp not in seen_state_pps:
                state_pairs.append((state, random.choice(prove_states), -1))
                num_negative += 1
                if num_negative >= num_positive * negative_ratio:
                    break
    return state_pairs


theorem = Theorem(repo, "MIL/C02_Basics/solutions/Solutions_S04_More_on_Order_and_Divisibility.lean",
                  "C02S04.Solutions_S04_More_on_Order_and_Divisibility_ex1")

state_pairs = \
    get_state_provability_data(theorem, tac_templates, 
                               theorem_code='Solutions_S04_More_on_Order_and_Divisibility_ex1 : max a b = max b a ',
                               max_steps = 100000,
                               verbose=True)

print(len(state_pairs))
```
