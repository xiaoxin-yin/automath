```python
import random
import sympy as sp

template = "c_0*f_0(c_1*t)+c_2"

functions = ['sin', 'cos', 'exp']

def random_constant():
    if random.random() < 0.33:
        return 1
    while True:
        if random.random() < 0.5:
            val = random.randint(-5, 5)
        else:
            val = round(random.uniform(-5.0, 5.0), 1)
        return val

def generate_non_poly_expr():
    func = random.choice(functions)
    while True:
        c_0 = random_constant()
        if c_0 != 0:
            break
    c_1 = random_constant()
    if random.random() < 0.33:
        c_2 = random_constant()
    else:
        c_2 = 0
    expr = template.replace('f_0', str(func)).replace('c_0', str(c_0)).replace('c_1', str(c_1)).replace('c_2', str(c_2))
    return sp.simplify(sp.sympify(expr))


def generate_poly_expr(max_degree=4):
    degree = random.randint(1, max_degree)
    t = sp.Symbol('t')
    while True:
        output = sp.sympify(0)
        for i in range(degree+1):
            if random.random() < 0.33:
                continue
            output += random_constant() * (t**i)
        if not output.is_constant():
            return output
    return None

fout = open("datasets/random_exprs_1m.txt", 'w')

for i in range(1000000):
    if i % 1000 == 0:
        print(i, "generated")
        fout.flush()
    if random.random() < 0.5:
        fout.write(str(generate_non_poly_expr()) + '\n')
    else:
        fout.write(str(generate_poly_expr()) + '\n')
    
fout.close()
```
